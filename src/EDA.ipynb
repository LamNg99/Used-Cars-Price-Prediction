## Data preprocessing 

```python
# drop features that won't be relevant in the analysis
df.drop(columns=['id', 'url', 'region_url', 'VIN', 'image_url', 'description', 'county'] , axis=1, inplace=True)
df.head()
```

```python
# Check for duplicated data
df.duplicated().value_counts()
```

```python
# Drop duplicated data 
df.drop_duplicates(inplace=True)
df.shape
```

```python
# View na value counts of each column
fig, ax = plt.subplots(figsize =(10, 5))
ax.bar(df.columns, df.isna().sum())
ax.set_ylabel('Missing values')
ax.set_ylim(0,df.shape[0])
ax.tick_params('x',labelrotation=90)
```

```python
# Drop na values in posting_date and change to datetime format
df=df.dropna(subset=['posting_date'],axis=0)
df['postdate'] = pd.to_datetime(df['posting_date'], utc=True)
df.postdate
```

```python
# Add datetime parameters for saledate column
df['postYear'] = df.postdate.dt.year
df['postMonth'] = df.postdate.dt.month
df['postDay'] = df.postdate.dt.day
```

```python
def remove_outliers(df, column_name):
    """
    Calculates and removes outliers from a pandas dataframe based on the interquartile range (IQR) method.

    Args:
        df (pandas.DataFrame): The dataframe to remove outliers from.
        column_name (str): The name of the column to remove outliers from.

    Returns:
        pandas.DataFrame: The dataframe with outliers removed.
    """
    q1 = df[column_name].quantile(0.25)
    q3 = df[column_name].quantile(0.75)
    iqr = q3 - q1
    lower_bound = q1 - 1.5 * iqr
    upper_bound = q3 + 1.5 * iqr
    df = df[(df[column_name] > lower_bound) & (df[column_name] < upper_bound)]
    return df
```
